```python
# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score


# Dropping Person ID
#df = df.drop('Person ID', axis=1)

#Converting String to Float
fraction = "120/80"

# Splitting the string at the '/'
numerator, denominator = fraction.split('/')

# Converting both parts to integers and then divide to get the float
result = float(numerator) / float(denominator)

# Encoding categorical variables (Gender, Occupation, BMI Category, Sleep Disorder)
label_encoders = {}
for column in ['Gender', 'Occupation', 'BMI Category', 'Sleep Disorder']:
    le = LabelEncoder()
    df[column] = le.fit_transform(df[column])
    label_encoders[column] = le

# Splitting features (X) and target (y)
X = df.drop('Sleep Disorder', axis=1)  # Features
y = df['Sleep Disorder']  # Target variable

# Splitting data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initializing classifiers
dt_model = DecisionTreeClassifier(random_state=42)
rf_model = RandomForestClassifier(random_state=42)

# Trainning models
dt_model.fit(X_train, y_train)
rf_model.fit(X_train, y_train)

# Predicting on test data
y_pred_dt = dt_model.predict(X_test)
y_pred_rf = rf_model.predict(X_test)

# Evaluating the models
print("Decision Tree Model")
print("Accuracy:", accuracy_score(y_test, y_pred_dt))
print("Classification Report:\n", classification_report(y_test, y_pred_dt))

print("Random Forest Model")
print("Accuracy:", accuracy_score(y_test, y_pred_rf))
print("Classification Report:\n", classification_report(y_test, y_pred_rf))
```
